# Load Image

```python
import matplotlib.pyplot as plt

import cv2
```

```python
filename = 'data/img01.jpg'
```

```python
image = cv2.imread(filename)
image.shape
```

```python
plt.imshow(image)
plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```
